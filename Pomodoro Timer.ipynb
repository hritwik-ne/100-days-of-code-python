```python
from datetime import datetime
from datetime import timedelta

def start(cycles, x, y, z):  
    task_count = 1
    goal = cycles
    task_length = timedelta(minutes=x)
    short_break = timedelta(minutes=y)
    long_break = timedelta(minutes=z)
    
    # Iterating through body until task count reaches our goal
    while task_count <= goal:
        print("* Cycle {} *".format(task_count))
        
        # Task timer
        start_time = datetime.now()
        print("Started: ({})".format(start_time.strftime('%H:%M:%S')))
        while True:
            now = datetime.now()
            if (now-start_time) >= task_length:
                print('Completed: ({})'.format(now.strftime('%H:%M:%S')))
                print('-------------')
                break
        
        # To commence break if the final cycle isn't over
        if task_count!=(goal):
            
            # Long break (After every 4 cycles)
            if (task_count%4) == 0:
                start_time = datetime.now()
                print("Long break started ({})".format(start_time.strftime('%H:%M:%S')))
                while True:
                    now = datetime.now()
                    if (now-start_time) >= long_break:
                        print("Long break ended ({})".format(now.strftime('%H:%M:%S')))
                        print("----------------")
                        break
            
            # Short break timer (default)
            else:
                start_time = datetime.now()
                print("Short break started ({})".format(start_time.strftime('%H:%M:%S')))
                while True:
                    now = datetime.now()
                    if (now-start_time) >= short_break:
                        print("Short break ended ({})".format(now.strftime('%H:%M:%S')))
                        print('-------------')
                        break
            
        task_count+=1
                            
            
def pomodoro():
    print('Pomodoro Timer\n\n1. Start\n2. Exit')
    print('-----------------')
    choice = input('Enter an option: ')
    
    if choice == '1':
        cycles = int(input('How many cycles: '))
        cl = int(input('Length of each cycle (in minutes): '))
        sb = int(input('Short Break length (in minutes): '))
        lb = int(input('Long Break length (in minutes): '))
        print('-----------------')
        start(cycles, cl, sb, lb)
        
    if choice == '2': exit
    else: exit  
```

```python
pomodoro()  # Run this cell for activating the timer
```
